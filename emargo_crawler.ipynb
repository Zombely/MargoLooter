```python
import requests
from tqdm import tqdm, trange
import time
from bs4 import BeautifulSoup
import re
from requests_html import AsyncHTMLSession

session = AsyncHTMLSession()
```

```python
MAIN_URL = "http://emargo.pl/"
```

```python
r = requests.get(f"{MAIN_URL}przedmioty/")
```

```python
soup = BeautifulSoup(r.text)
```

```python
a = ''
for element_a in soup.find_all("a", {"href": re.compile("\/przedmioty\/dla.*")})[:1]:
    href = MAIN_URL[:-1] + element_a['href']
    soup_items = BeautifulSoup(requests.get(href).text)
    items = soup_items.find_all("a", {"href": re.compile("\/przedmiot\/.*")})
    for item in items[:1]:
        item_h =MAIN_URL[:-1] + item['href']
        print(item_h)
        s_r = await session.get(item_h)
        await s_r.html.arender()
        
        # soup_item = BeautifulSoup(.text)
        # item_info = soup_item.find_all("div", {"id": "itemInfo"})
        a=s_r.html
```

```python
item_data = a.find("#infoItem",first=True)
item_info = {
    "name": item_data.find("b", first=True).text,
    "data": item_data.text.replace(item_data.find("b", first=True).text, ""),
    "image_href": MAIN_URL[:-1]+a.find("img", first=True).attrs['src']
}
```

```python
for element in a.find(".col-3-2", first=True).find(".col-2"):
    item_info[element.text.split("\n")[0]] = element.text.split("\n")[1]
```
